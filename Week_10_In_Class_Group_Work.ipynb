Computational models, including machine learning models, require numerial input variables. However, in the world of data collection, some variables are not inherently numerical. Variables for attributes such as name, color, gender, degree of satisfaction, etc., are categorical variables. While there are two types of categorical data, ordinal and nominal, ordinal data can be ranked, and is easily converted to numerical data using ordinal encoding. Nominal categorial data, however, by its nature has no inherent or natural ordering, and so it must be artificially transformed. Enter "one-hot coding". Essentially, we create a column for each particular state or value of a nominal variable and insert it in the dataframe. Then, for each entry or row, we record whether that particular state or value is true (1) or false (0). No row can have more than one state for a given variable. 

Below, I illustrate an example of one-hot encoding using data from the Belize Airbnb dataframe, using the variable "neighbourhood groups".
